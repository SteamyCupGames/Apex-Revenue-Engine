# 01 EDA Cleaning

This analysis uses the Olist dataset to address a critical marketing challenge: user retention. The goal is to identify high-value customer segments  and predict the risk of churn in order to optimize campaign ROI.

```python
import pandas as pd

# Load Key Datasets
orders = pd.read_csv('../Data/Raw/olist_orders_dataset.csv')
payments = pd.read_csv('../Data/Raw/olist_order_payments_dataset.csv')
customers = pd.read_csv('../Data/Raw/olist_customers_dataset.csv')
```

```python
# 1. Join orders with payments (an order can have multiple payment methods)
df_orders_pay = pd.merge(orders, payments, on='order_id', how='left')

# 2. Join with customers to have the 'customer_unique_id'
# (Note: customer_id is session-based, customer_unique_id is the actual person)
master_df = pd.merge(df_orders_pay, customers, on='customer_id', how='inner')
```

It was decided to use an inner join between the orders and customers tables to ensure data integrity, filtering out records without a valid customer_unique_id, which account for less than 1% of the dataset.

```python
# 3. Convert dates to datetime
date_cols = ['order_purchase_timestamp', 'order_delivered_customer_date']
for col in date_cols:
    master_df[col] = pd.to_datetime(master_df[col])

# Save progress
master_df.to_csv('../Data/Processed/master_cleaned.csv', index=False)
```
